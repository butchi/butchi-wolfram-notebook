```
f1p[x_] := x ** Quaternion[1, 0, 0, 0];
f1m[x_] := -x ** Quaternion[1, 0, 0, 0] + Quaternion[1, 0, 0, 0];
f2p[x_] := x ** Quaternion[0, 1, 0, 0] + Quaternion[1, 0, 0, 0];
f2m[x_] := -x ** Quaternion[0, 1, 0, 0] + Quaternion[1, 1, 0, 0];
f3p[x_] := x ** Quaternion[0, 0, 1, 0] + Quaternion[1, 1, 0, 0];
f3m[x_] := -x ** Quaternion[0, 0, 1, 0] + Quaternion[1, 1, 1, 0];
```

```
ifs = {f1p[#], f2p[#], f3p[#]} &;
init = {{0, 1}};
Graphics3D[{Table[
    Arrow[PadRight[#, 3] & /@ (List @@@ item)], {
        item, 
        Nest[Flatten[ifs[#], 1] &, init, 1]
    }
]}]
Graphics3D[{Table[
    Line[PadRight[#, 3] & /@ (List @@@ item)], {
        item, 
        Nest[Flatten[ifs[#], 1] &, init, 5]
    }
]}]
```
