# bion-waon

date: 2022-09-24

```
waon[n_, b_] := Exp[2 Pi Log[n, b] I]
waon[n_] := Exp[2 Pi Log[n] I]
waon10[n_] := Exp[2 Pi Log10[n] I]
waon2[n_] := Exp[2 Pi Log2[n] I]
```

```
bion[n_, b_] := Log[n, b] waon[n, b]
bion[n_] := Log[n] waon[n]
bion10[n_] := Log10[n] waon10[n]
bion2[n_] := Log2[n] waon2[n]
```

```
With[
    {n = 256}, 
    Graphics[
        Table[{
            Hue[Log2[3] + Log2[k], 1 - (Log2[k]/Log2[n])^2, 1], 
            Disk[ReIm[waon2[k]], 0.3 k^-0.2]
        }, {k, n, 1, -2}]
    ]
]
```

```
With[
    {n = 256}, 
    Graphics[
        Reverse[Table[{
            Hue[Log2[3] + Log2[k], 1 - (Log2[k]/Log2[n])^2, 1], 
            Disk[ReIm[waon2[k]], 0.3 k^-0.2],
            Black,
            Text[ToString[k], ReIm[waon2[k]]]
        }, {k, n}]]
    ]
]
```

```
With[
    {n = 2047}, 
    Graphics[
        Table[{
            Hue[Log2[3], (1 - Log2[k]/Log2[n])^2, 1], 
            Disk[ReIm[bion2[k]], 0.5 k^-0.25]
        }, {k, n, 1, -1}]
    ]
]
```
