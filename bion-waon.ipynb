```
waon[b_, n_] := Exp[2 Pi Log[b, n] I]
waon[n_] := Exp[2 Pi Log[n] I]
waon10[n_] := Exp[2 Pi Log10[n] I]
waon2[n_] := Exp[2 Pi Log2[n] I]
```

```
bion[b_, n_] := Log[b, n] waon[b, n]
bion[n_] := Log[n] waon[n]
bion10[n_] := Log10[n] waon10[n]
bion2[n_] := Log2[n] waon2[n]
```

```
integerOdd[n_Integer] := n / (2 ^ IntegerExponent[n, 2])
```

```
bionVolume[x_, y_] := integerOdd[LCM[x, y] / GCD[x, y]]
```

```
bionDistance[x_, y_] := Log2[bionVolume[x, y]]
```

```
n1 = 2 * 3 * 4 *  5 *  6
n2 = 7 * 8 * 9 * 10 * 11
bionVolume[n1, n2]
N[bionDistance[n1, n2]]
```
