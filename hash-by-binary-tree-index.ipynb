# hash-by-binary-tree-index.nb

from binary-tree-index.nb

created at 2021-10-17

```
indexReal[n_] := FromDigits[{RotateLeft[IntegerDigits[n, 2]], 0}, 2]
```

```
hash[n_, bit_] := Floor[2^bit indexReal[n]]
```

```
tbl = Table[hash[n, 8], {n, 0, 2^8 - 1}]
ListPlot[tbl,
    PlotRange -> {{0, 256}, {0, 256}}, 
    AspectRatio -> Automatic
]
DuplicateFreeQ[tbl]
Length[Union[tbl]]
```

```
tbl = Table[hash[n, 8], {n, 0, 2^10 - 1}]
ListPlot[tbl, PlotRange -> {{0, 1024}, {0, 256}}]
DuplicateFreeQ[tbl]
Length[Union[tbl]]
```
