# Fibonacci Kion Experiment

## フィボナッチ貝音

date: 2024-01-18

bionのフィボナッチ版を「貝音」(kion)と名付け、黄金螺旋との違いを示すとともに、整数螺旋（これも造語）として奇数の2べき倍の代わりにWythoff配列を用いて可視化しました。

```
ListLinePlot[
    ReIm /@ Table[GoldenRatio^(i/12) E^(I GoldenRatio i/12), {i, 0, 99}],
AspectRatio -> Automatic, PlotRange -> All]
```

```
ListLinePlot[
    Table[ReIm[Fibonacci[i] E^(2 Pi I i)], {i, 1, 7, 1/256}], 
AspectRatio -> Automatic, PlotRange -> All]
```

```
wythoff[n_, k_] := Fibonacci[k + 1] Floor[n*GoldenRatio] + (n - 1) Fibonacci[k];
Table[wythoff[n, k], {n, 10}, {k, 10}] // TableForm
```

```
Table[wythoff[i, 1], {i, 25}]
```

```
Show[{ParametricPlot[
    AngleVector[{Fibonacci[t], 2 \[Pi] t}], {t, 1, 11}, 
    PlotRange -> {{-63, 63}, {-63, 63}}, PlotStyle -> LightGray], 
    Graphics[{
        Table[
            With[
                {w = wythoff[j, i]},
                Text[w, AngleVector[{w, 2 Pi i + {0, 4.2, 3.0, 2.0, 5.3, 1.2, 3.8, 4.9, 0.7, 2.5, 3.5, 4.5, 5.8, 0.4, 1.5}[[j]]}]]
            ], {i, 15}, {j, 15}
        ]
    }]
}]
```

```
ListPlot[{0, 4.2, 3.0, 2.0, 5.3, 1.2, 3.8, 4.9, 0.7, 2.5, 3.5, 4.5, 5.8, 0.4, 1.5}]
```

```
ListPlot[{4.2, 3.0 + 2 Pi, 2.0 + 4 Pi, 5.3 + 4 Pi, 1.2 + 6 Pi, 3.8 + 6 Pi, 4.9 + 6 Pi, 0.7 + 8 Pi, 2.5 + 8 Pi, 3.5 + 8 Pi, 4.5 + 8 Pi, 5.8 + 8 Pi, 0.4 + 10 Pi, 1.5 + 10 Pi}]
```

```
(* 合わない *)
Show[{ParametricPlot[
    AngleVector[{Fibonacci[t], 2 \[Pi] t}], {t, 1, 11}, 
    PlotRange -> {{-63, 63}, {-63, 63}}, PlotStyle -> LightGray], 
    Graphics[{
        Table[
            With[{w = wythoff[j, i]}, 
            Text[w, AngleVector[{w, 
            2 Pi i + 2 Pi Log[GoldenRatio, j] - Pi}]]], {i, 15}, {j, 15}]
    }]
}]
```
