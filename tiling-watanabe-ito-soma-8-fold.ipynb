```
rectG = {Orange, EdgeForm[Thick], Rectangle[{0, 0}, {1, 1}]};
arrowG = {
    LightBlue,
    EdgeForm[Thick],
    Polygon[{{0, 0}, {1/Sqrt[2], 1/Sqrt[2]}, {1/Sqrt[2], 1/Sqrt[2] + 1}, {0, 1}}],
    Black,
    Disk[0.85 {1/Sqrt[2], 1/Sqrt[2] + 1}, 0.015]
};

{Graphics[rectG], Graphics[arrowG]}
```

```
{
    Graphics[
        GeometricTransformation[{
            GeometricTransformation[arrowG, Table[RotationTransform[(2 Pi i)/8], {i, 0, 7}]],
            GeometricTransformation[rectG, Table[RightComposition[
                TranslationTransform[{1/Sqrt[2], 1/Sqrt[2]}],
                RotationTransform[(2 Pi i)/8]
            ], {i, 0, 7}]]
        }, RightComposition[
            TranslationTransform[{1 + 1/Sqrt[2], 1 + 1/Sqrt[2]}], ScalingTransform[{1/(2 + Sqrt[2]), 1/(2 + Sqrt[2])}]
        ]]
    , Axes -> True],

    Graphics[
        GeometricTransformation[{
            arrowG,
            GeometricTransformation[arrowG, TranslationTransform[{1, 1 + Sqrt[2]}]],
            GeometricTransformation[arrowG, TranslationTransform[{1 + 1/Sqrt[2], 2 + 3/Sqrt[2]}]],
            GeometricTransformation[arrowG, RightComposition[RotationTransform[-Pi/2], TranslationTransform[{0, 1 + Sqrt[2]}]]],
            GeometricTransformation[arrowG, RightComposition[RotationTransform[-Pi/4], TranslationTransform[{0, 2 + Sqrt[2]}]]],
            GeometricTransformation[arrowG, RightComposition[RotationTransform[Pi/4], TranslationTransform[{1 + Sqrt[2], 1 + Sqrt[2]}]]],
            GeometricTransformation[rectG, TranslationTransform[{1/Sqrt[2], 1/Sqrt[2]}]],
            GeometricTransformation[rectG, RightComposition[TranslationTransform[{1/Sqrt[2], 1/Sqrt[2]}], RotationTransform[Pi/4]]],
            GeometricTransformation[rectG, RightComposition[RotationTransform[Pi/4], TranslationTransform[{1 + 1/Sqrt[2], 1 + 1/Sqrt[2]}]]],
            GeometricTransformation[rectG, RightComposition[TranslationTransform[{0, 1 + Sqrt[2]}]]],
            GeometricTransformation[rectG, RightComposition[RotationTransform[Pi/4], TranslationTransform[{1 + Sqrt[2], 2 + Sqrt[2]}]]],
            GeometricTransformation[rectG, RightComposition[TranslationTransform[{1/Sqrt[2], 2 + 3/Sqrt[2]}]]]
        }, ScalingTransform[{1/(2 + Sqrt[2]), 1/(2 + Sqrt[2])}]],
    Axes -> True]
 }
```

```
Graphics[Nest[# /. {
    rectKey -> GeometricTransformation[{
        GeometricTransformation[arrowKey, Table[RotationTransform[(2 Pi i)/8], {i, 0, 7}]],
        GeometricTransformation[rectKey, Table[RightComposition[
            TranslationTransform[{1/Sqrt[2], 1/Sqrt[2]}],
            RotationTransform[(2 Pi i)/8]
        ], {i, 0, 7}]]
    }, RightComposition[
        TranslationTransform[{1 + 1/Sqrt[2], 1 + 1/Sqrt[2]}], ScalingTransform[{1/(2 + Sqrt[2]), 1/(2 + Sqrt[2])}]
    ]],
    arrowKey -> GeometricTransformation[{
        arrowKey,
        GeometricTransformation[arrowKey, TranslationTransform[{1, 1 + Sqrt[2]}]],
        GeometricTransformation[arrowKey, TranslationTransform[{1 + 1/Sqrt[2], 2 + 3/Sqrt[2]}]],
        GeometricTransformation[arrowKey, RightComposition[RotationTransform[-Pi/2], TranslationTransform[{0, 1 + Sqrt[2]}]]],
        GeometricTransformation[arrowKey, RightComposition[RotationTransform[-Pi/4], TranslationTransform[{0, 2 + Sqrt[2]}]]],
        GeometricTransformation[arrowKey, RightComposition[RotationTransform[Pi/4], TranslationTransform[{1 + Sqrt[2], 1 + Sqrt[2]}]]],
        GeometricTransformation[rectKey, TranslationTransform[{1/Sqrt[2], 1/Sqrt[2]}]],
        GeometricTransformation[rectKey, RightComposition[TranslationTransform[{1/Sqrt[2], 1/Sqrt[2]}], RotationTransform[Pi/4]]],
        GeometricTransformation[rectKey, RightComposition[RotationTransform[Pi/4], TranslationTransform[{1 + 1/Sqrt[2], 1 + 1/Sqrt[2]}]]],
        GeometricTransformation[rectKey, RightComposition[TranslationTransform[{0, 1 + Sqrt[2]}]]],
        GeometricTransformation[rectKey, RightComposition[RotationTransform[Pi/4], TranslationTransform[{1 + Sqrt[2], 2 + Sqrt[2]}]]],
        GeometricTransformation[rectKey, RightComposition[TranslationTransform[{1/Sqrt[2], 2 + 3/Sqrt[2]}]]]
    }, ScalingTransform[{1/(2 + Sqrt[2]), 1/(2 + Sqrt[2])}]]
} &, {rectKey}, 3] /. {
    rectKey -> {
        Orange,
        EdgeForm[Thin],
        Rectangle[{0, 0}, {1, 1}]
    },
    arrowKey -> {
        LightBlue,
        EdgeForm[Thin],
        Polygon[{{0, 0}, {1/Sqrt[2], 1/Sqrt[2]}, {1/Sqrt[2], 1/Sqrt[2] + 1}, {0, 1}}]
    }
}]
```
