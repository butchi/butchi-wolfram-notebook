# PrimeQK Prime List

素数大富豪用の素数リスト

```
primeDig3 = Select[
    Range[99 + 1, 999],
    PrimeQ
]
```

```
FactorInteger[Length[primeDig3]]
```

```
ExportString[Partition[primeDig3, 11], "Table"]
```

```
cardNumLi = Range[1, 13]
```

```
cardStrLi = StringReplace[IntegerString[#, 36], {"1" -> "A", "a" -> "T", "b" -> "J", "c" -> "Q", "d" -> "K"}] & /@ cardNumLi
```

```
parseCardString[str_] := FromDigits[StringReplace[str, {"A" -> "1", "T" -> "10", "J" -> "11", "Q" -> "12", "K" -> "13"}]]
```

```
primeCard3 = SortBy[Select[StringJoin @@@ Tuples[cardStrLi, 3], With[{num = parseCardString[#]}, num > 999 && PrimeQ[num]]&], parseCardString]
```

```
Length[primeCard3]
FactorInteger[Length[primeCard3]]
Divisors[Length[primeCard3]]
```

```
14 ^ 2
```

```
ExportString[Partition[primeCard3, 14], "Table"]
```
