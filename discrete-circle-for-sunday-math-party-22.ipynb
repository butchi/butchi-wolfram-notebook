```
exportPng[expr_] := "data:image/png;base64," <> ExportString[ExportString[expr, "PNG"], "Base64"]
exportGif[expr_] := "data:image/gif;base64," <> ExportString[ExportString[expr, "GIF"], "Base64"]
```

```
ArrayPlot[
    Table[Mod[Floor[Sqrt[x^2 + y^2]], 2], {y, -256, 256}, {x, -256, 256}],
    PixelConstrained -> 1,
    Frame -> False
]
```

```
ArrayPlot[
    RotateRight[
        Log@Abs[Fourier[
            Table[Mod[Floor[Sqrt[x^2 + y^2]], 2], {y, -256, 256}, {x, -256, 256}]
        ]],
        {256, 256}
    ],
    PixelConstrained -> 1,
    Frame -> False
]
```

```
discLi = Table[
    ArrayPlot[
        Table[HeavisidePi[Sqrt[x^2 + y^2]/r], {y, -64, 63}, {x, -64, 63}],
        PixelConstrained -> 4
    ],
    {r, 63}
]
CopyToClipboard[exportGif[discLi]]
```

```
circleLi = Table[
    ArrayPlot[
        Table[
            HeavisidePi[Sqrt[x^2 + y^2]/r] - HeavisidePi[Sqrt[x^2 + y^2]/(r + 2)],
            {y, -64, 63},
            {x, -64, 63}
        ],
        PixelConstrained -> 4
    ],
    {r, 63}
]
CopyToClipboard[exportGif[circleLi]]
```

```
altLi = Table[
    ArrayPlot[
        Table[HeavisidePi[Sqrt[x^2 + y^2]/r] Mod[Floor[Sqrt[x^2 + y^2]], 2], {y, -64, 63}, {x, -64, 63}],
        PixelConstrained -> 4
    ],
    {r, 63}
]
CopyToClipboard[exportGif[altLi]]
```
