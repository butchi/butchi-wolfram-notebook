# Square root of 1

自己同形数 (Automorphic number)

created-at: 2024-02-17

```
n = 32;
```

```
a = PowerMod[5, 2^n, 10^n]
b = PowerMod[16, 5^n, 10^n]
```

```
Select[Range[9999], Mod[#^2, 10000] == # &]
```

```
Mod[a^2, 10^n]
Mod[b^2, 10^n]
```

```
Mod[a + b, 10^n]
Mod[a b, 10^n]
```

```
c = 2 a - 1
d = 2 b - 1
```

```
e = 10^n - a
f = 10^n - b
```
