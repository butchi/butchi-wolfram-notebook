```
e = E^((2 \[Pi] I)/8)
```

```
FullSimplify[e]
```

```
ePowLi = Table[FromDigits[li, e], {li, Tuples[Range[0, 1], 8]}];
```

```
Length[ePowLi]
```

```
Union[N[ePowLi]] // Length
```

```
ListPlot[ReIm /@ Union[N[ePowLi]], AspectRatio -> Automatic]
```

```
ListPlot[ReIm /@ 
    Flatten[Table[N[ePowLi + x + y I], {y, -3, 3}, {x, -3, 3}]], 
    AspectRatio -> Automatic, PlotRange -> {{-3, 3}, {-3, 3}}
]
```

```
ListPlot[ReIm /@ 
  Flatten[Table[N[ePowLi + x + y I], {y, -3, 3}, {x, -3, 3}]], 
  AspectRatio -> Automatic, PlotRange -> {{0, 1}, {0, 1}}
]
```

```
t0 = Sqrt[2] - 1;

t1 = (1 - Sqrt[2]/2) + (1 - Sqrt[2]/2) I;

rx[x_] := (1 - Re[x]) + Im[x] I

ry[x_] := Re[x] + (1 - Im[x]) I

rxy[x_] := (1 - Re[x]) + (1 - Im[x]) I

tr[x_] := Im[x] + Re[x] I
```

```
ListPlot[ReIm /@ 
    Flatten[Table[{rf[0], rf[t0], rf[tr[t0]], rf[t1]}, {rf, {Identity, rx, ry, rxy}}]], 
    AspectRatio -> Automatic
]
```

```
Graphics[{Table[
    Point[ReIm[pt] + {x, y}], {y, -3, 3}, {x, -3, 3}, {
        pt, 
        Union[Flatten[
            Table[rf[pt], {rf, {Identity, rx, ry, rxy}}, {pt, {0, t0, tr[t0], t1}}]
        ]]
    }]},
    PlotRange -> {{-1.5, 1.5}, {-1.5, 1.5}}, 
    ImageSize -> Large
]
```

```
nf = Nearest[
  Join @@ Join @@ 
  Table[ReIm[pt] + {x, y}, {y, -1, 1}, {x, -1, 1}, {
    pt, 
    Union[Flatten[
      Table[rf[pt], {rf, {Identity, rx, ry, rxy}}, {pt, {0, t0, tr[t0], t1}}]
    ]]
  }]
]
```

```
DensityPlot[
    First[nf[{x, y}]], {x, -(1/2), 1/2}, {y, -(1/2), 1/2}, PlotPoints -> 100, ColorFunction -> GrayLevel
]
```
