```
exportMp3[expr_] := "data:audio/mpeg;base64," <> ExportString[ExportString[expr, "MP3"], "Base64"]
```

```
arrayPlot3d[li_] := Graphics3D[{
    With[
        {dim = Dimensions[li]}, 
        Table[
            If[li[[x, y, z]] == 1, Cuboid[{x, y, z}], {}],
            {z, dim[[1]]},
            {y, dim[[2]]},
            {x, dim[[3]]}
        ]
    ]
}]
```

```
arrayPlot3d[Nest[ArrayFlatten[# /. x_ -> {{
    {x, x, x},
    {x, 0, x},
    {x, x, x}
} , {
    {x, 0, x},
    {0, 0, 0},
    {x, 0, x}
} , {
    {x, x, x},
    {x, 0, x},
    {x, x, x}
}}, 3] &, 1, 4]]
```

```
snd = ListPlay[Flatten[Nest[ArrayFlatten[# /. x_ -> {{
    {x, x, x},
    {x, 0, x},
    {x, x, x}
} , {
    {x, 0, x},
    {0, 0, 0},
    {x, 0, x}
} , {
    {x, x, x},
    {x, 0, x},
    {x, x, x}
}}, 3] &, 1, 4]]];

Short[snd]

CopyToClipboard[exportMp3[snd]]
```

```
li = Nest[ArrayFlatten[# /. x_ -> {{
    {0, 0},
    {0, x}
} , {
    {0, x},
    {x, x}
}}, 3] &, 1, 6];

Graphics3D[{
    With[{dim = Dimensions[li]}, 
        Table[
            If[li[[x, y, z]] == 1, Cuboid[{x, y, z}], Null],
            {z, dim[[1]]},
            {y, dim[[2]]},
            {x, dim[[3]]}
        ]
    ]
}]

snd = ListPlay[Flatten[li], SampleRate -> 8000];

Short[snd]

CopyToClipboard[exportMp3[snd]]
```
