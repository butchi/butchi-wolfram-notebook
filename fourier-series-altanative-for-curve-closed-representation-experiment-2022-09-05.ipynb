# fourier-series-altanative-for-curve-closed-representation-experiment-2022-09-05.ipynb

created at 2022-09-05


```
SeedRandom[1];
f[t_] := Total[a E^(s I Range[n] t)];
n = 5;
a = (RandomReal[1, n] E^(I RandomReal[2 \[Pi], n])) / Range[n]^2;
s = (-1)^RandomInteger[1, n];
ListLogPlot[
    Abs[a]*# & /@ Boole[{Positive[s], Negative[s]}], 
    Filling -> Axis, PlotRange -> All, 
    PlotStyle -> {
        { Blue, PointSize[Large] },
        { Red, PointSize[Large] }
    }
]
ParametricPlot[
    {Re[#], Im[#]} &[f[t]],
    {t, 0, 2 \[Pi]}, 
    PlotRange -> All
]
```
