```
(* 九蓮宝燈の多項式 *)
Expand[FromDigits[Reverse[{4, 1, 1, 1, 1, 1, 1, 1, 3}], x]]
```

```
(* 両面塔子 *)
Table[Expand[(1 + x) x^n], {n, 8}]
```

```
(* 順子 *)
Table[Expand[(1 + x + x^2) x^n], {n, 7}]
```

```
(* 刻子 *)
Table[Expand[3 x^n], {n, 9}]
```

```
(* 4面子1雀頭の分割多項式 *)
Prepend[#, 2] & /@ 
    Union[
        Sort /@ Permutations[{3, 3, 3, 3, x + x^2 + x^3, x + x^2 + x^3, x + x^2 + x^3, x + x^2 + x^3}, {4}]
    ]
```

```
(* 孤立牌がない場合のブロック *)
Union[Sort /@ 
  Select[Permutations[
    Join[Table[2, 7], Table[x + x^2, 4], Table[3, 4], 
     Table[x + x^2 + x^3, 4]], 7], (Total[#] /. x -> 1) == 14 &]]
```

```
(* 14牌の分割数（5ブロック(4面子1雀頭)～14ブロック(全部孤立牌)） *)
IntegerPartitions[14, All, 3]
```

```
(* 試しに整数ブロックを多項式に置換してみる *)
Tuples[{3, 3, 2, 1, 1, 1, 1, 1, 1} /. {1 -> {1}, 2 -> {2, 1 + x}, 3 -> {3, 1 + x + x^2}}]
```

```
Union[Sort /@ (Join @@ (
    Sort /@ Table[
        Tuples[ipLi /. {1 -> {1}, 2 -> {2, 1 + x, 1 + x^2}, 3 -> {3, 1 + x + x^2}}],
    {ipLi, IntegerPartitions[14, All, 3]}]
))]
```
